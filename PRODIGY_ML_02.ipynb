```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data = pd.read_csv("customer_1.csv")

data.drop(["Customer ID", "Purchase Date", "Customer Name"], axis=1, inplace=True)

data.fillna(0, inplace=True)  

data = pd.get_dummies(data, columns=["Product Category", "Payment Method", "Gender"])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

kmeans = KMeans(n_clusters=3, random_state=42)  
kmeans.fit(scaled_data)
data['Cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
plt.scatter(data['Total Purchase Amount'], data['Customer Age'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Customer Age')
plt.title('K-means Clustering of Customers')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = pd.read_csv("customer_1.csv")

data.drop(["Customer ID", "Purchase Date", "Customer Name"], axis=1, inplace=True)

data.fillna(0, inplace=True)  
data = pd.get_dummies(data, columns=["Product Category", "Payment Method", "Gender"])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

kmeans = KMeans(n_clusters=3, random_state=42)  
kmeans.fit(pca_data)
data['Cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=data['Cluster'], cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-means Clustering of Customers')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data = pd.read_csv("customer_1.csv")

data.drop(["Customer ID", "Purchase Date", "Customer Name"], axis=1, inplace=True)

data.fillna(0, inplace=True)  

data = pd.get_dummies(data, columns=["Product Category", "Payment Method", "Gender"])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

kmeans = KMeans(n_clusters=3, random_state=42) 
kmeans.fit(scaled_data)
data['Cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
scatter = plt.scatter(data["Total Purchase Amount"], data["Customer Age"], c=data['Cluster'], cmap='viridis')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Customer Age')
plt.title('K-means Clustering of Customers')
plt.colorbar(label='Cluster')

unique_clusters = data['Cluster'].unique()
handles = [plt.Line2D([0], [0], marker='o', color='w', label=f'Cluster {i}', 
                      markerfacecolor=plt.cm.viridis(i / len(unique_clusters)), markersize=10) for i in unique_clusters]
plt.legend(handles=handles, title="Clusters")
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import plotly.express as px

data = pd.read_csv("customer_1.csv")
data.drop(["Customer ID", "Purchase Date", "Customer Name"], axis=1, inplace=True)
data.fillna(0, inplace=True)  
data = pd.get_dummies(data, columns=["Product Category", "Payment Method", "Gender"])

selected_features = ["Total Purchase Amount", "Quantity"]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[selected_features])

kmeans = KMeans(n_clusters=3, random_state=42)  
kmeans.fit(scaled_data)
data['Cluster'] = kmeans.labels_

sampled_data = data.sample(frac=0.1, random_state=42)  

fig = px.scatter(
    sampled_data, 
    x="Total Purchase Amount", 
    y="Quantity", 
    color="Cluster", 
    title="K-means Clustering of Customers",
    labels={
        "Total Purchase Amount": "Total Purchase Amount",
        "Quantity": "Quantity",
        "Cluster": "Cluster"
    }
)

fig.show()
```
